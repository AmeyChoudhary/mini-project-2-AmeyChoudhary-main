```python
# ChatGPT
# open t.txt 
import matplotlib.pyplot as plt

# Read the recorded timeline data from the file
timeline_data = []
with open("t.txt", "r") as file:
    for line in file:
        pid, queue_level, ticks = map(int, line.strip().split())
        timeline_data.append((ticks, pid, queue_level))


# Create a dictionary to store data for each PID
pid_data = {}
for ticks, pid, queue_level in timeline_data:
    if pid not in pid_data:
        pid_data[pid] = {"ticks": [], "queue_level": []}
        # pid_data[pid]["ticks"].append(0) uncomment this to show from 0
        # pid_data[pid]["queue_level"].append(0)

    pid_data[pid]["ticks"].append(ticks)
    pid_data[pid]["queue_level"].append(queue_level)

# Plot queue levels vs. time for each PID
for pid, data in pid_data.items():
    plt.plot(data["ticks"], data["queue_level"], label=f"PID {pid}")

# Customize the plot
# increase the size
plt.rcParams['figure.figsize'] = [15,10]
plt.xlabel("Time (ticks)")
plt.ylabel("Queue Level")
plt.title("Queue Levels vs. Time for Different PIDs")
plt.legend()
plt.grid(True)

# Show the plot or save it to a file
plt.show()
# You can save the plot to a file with plt.savefig("queue_levels_vs_time.png") if needed
```
